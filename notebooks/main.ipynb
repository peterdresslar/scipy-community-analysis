## SciPy and its Evolving Role in Science

### 1. About SciPy

This notebook is an attempt to explore the history of the Python scientific computing library (and community) [SciPy](https://scipy.org) from the perspective of the fields of study it serves and perhaps helps to shape.

Scipy is both a computational [software repository](https://github.com/scipy/scipy), based on Github, as well as a social organization with a [governance document](https://docs.scipy.org/doc/scipy/dev/governance.html) and a Steering Council and a [Benevolent Dictator For Life](https://docs.scipy.org/doc/scipy/dev/governance.html#bdfl), Pauli Virtanen.

SciPy's history is extensive, extending back to the early days of the organized Internet. *The State of SciPy* (Millman & Vaught, 2008) tracks the history of the SciPy software project to the creation of a Python Matrix object in 1994. (p. 5.) The organizational governance document, formalizing institutional control of the community, was initialized in July, 2017.

## 3. SciPy and Scientific Areas

SciPy's increase in general awareness, use, and development, might be thought of as not only a vertical increase, but a horizontal one. The fact that more researchers are using SciPy is not only a result of more users doing the same types of analysis, but rather the spread of SciPy into new disciplines where computational science can be adapted. 

One way we can investigate this is through the investigation of various issues and pull requests communicated through the Github repository. 

For instance, on August 21, 2018, a Github user named ChristianSauer communicated the following in a bug report, [Issue #9165](https://github.com/scipy/scipy/issues/9165):

> I am trying to use scipy.dendrogram to create a dendrogram of genomics data. Unfortunately, the dendrogram fails with a recursion error. THis code worked for slightly smaller datasets but maybe the data is malformed. [sic]
